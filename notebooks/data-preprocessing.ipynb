```python
import zipfile
import os
import pandas as pd
```

```python
zip_path = 'stock_data.zip'
extract_path = 'stock_data'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

files = os.listdir(extract_path)
print(files)
```

```python
#make an empty dataset called closed_prices
close_prices = pd.DataFrame()

for file in files:
    #extract files from the zip
    file_path = os.path.join(extract_path, file)
    #change the date column to date type
    df = pd.read_csv(file_path, parse_dates=['date'], index_col='date')
    #Make a stock name list by using names of each file to identify stocks
    stock_name = file.split('.')[0]  
    #make columns in close_prices and put values of close column in them
    close_prices[stock_name] = df['close']
#display the created file
print(close_prices.head())
#save the csv file for the next questions
close_prices.to_csv('combined_close_prices.csv')

```
